```python
####       Linear correlation       ######
from scipy.stats.stats import pearsonr
print(pearsonr(train_data.signal, train_data.open_channels))
#####      Remove drift     #######
def f(x,low,high,mid): return -((-low+high)/625)*(x-mid)**2+high -low
# Clean train batch 2
a=500000; b=600000 #
train_data.loc[train_data.index[a:b],'signal'] = train_data.signal[a:b].values - 3*(train_data.time.values[a:b] - 50)/10
# Clean train batch 7
batch = 7; a = 500000*(batch-1); b = 500000*batch
train_data.loc[train_data.index[a:b],'signal'] = train_data.signal.values[a:b] - f(train_data.time[a:b].values,-1.817,3.186,325)
# CLEAN TRAIN BATCH 8
batch = 8; a = 500000*(batch-1); b = 500000*batch
train_data.loc[train_data.index[a:b],'signal'] = train_data.signal.values[a:b] - f(train_data.time[a:b].values,-0.094,4.936,375)
# CLEAN TRAIN BATCH 9
batch = 9; a = 500000*(batch-1); b = 500000*batch
train_data.loc[train_data.index[a:b],'signal'] = train_data.signal.values[a:b] - f(train_data.time[a:b].values,1.715,6.689,425)
# CLEAN TRAIN BATCH 10
batch = 10; a = 500000*(batch-1); b = 500000*batch
train_data.loc[train_data.index[a:b],'signal'] = train_data.signal.values[a:b] - f(train_data.time[a:b].values,3.361,8.45,475)
```
